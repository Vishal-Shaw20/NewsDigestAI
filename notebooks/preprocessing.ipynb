```python
import pandas as pd
import re
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
data = pd.read_csv('../data/raw_news.csv')
data
```

```python
data_useful = data[['title','description','content']].copy()
data_useful
```

```python
data_useful.loc[:,'title'] = data_useful['title'].str.lower()
data_useful.loc[:,'description'] = data_useful['description'].str.lower()
data_useful.loc[:,'content'] = data_useful['content'].str.lower()
data_useful
```

```python
data_useful.loc[:,'title'] = data_useful['title'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', ' ', x))
data_useful.loc[:,'description'] = data_useful['description'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', ' ', x))
data_useful.loc[:,'content'] = data_useful['content'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', ' ', x))
data_useful
```

```python
data_useful.loc[:,'title'] = data_useful['title'].apply(lambda x: re.sub(r'\s+', ' ', x))
data_useful.loc[:,'description'] = data_useful['description'].apply(lambda x: re.sub(r'\s+', ' ', x))
data_useful.loc[:,'content'] = data_useful['content'].apply(lambda x: re.sub(r'\s+', ' ', x))
data_useful
```

```python
data_useful.loc[:,'title_tokens'] = data_useful['title'].apply(lambda x: word_tokenize(x))
data_useful.loc[:,'description_tokens'] = data_useful['description'].apply(lambda x: word_tokenize(x))
data_useful.loc[:,'content_tokens'] = data_useful['content'].apply(lambda x: word_tokenize(x))
data_useful
```

```python
stop_words = set(stopwords.words('english'))
```

```python
data_useful.loc[:,'title_tokens'] = data_useful['title_tokens'].apply(lambda tokens: [k for k in tokens if k not in stop_words])
data_useful.loc[:,'description_tokens'] = data_useful['description_tokens'].apply(lambda tokens: [k for k in tokens if k not in stop_words])
data_useful.loc[:,'content_tokens'] = data_useful['content_tokens'].apply(lambda tokens: [k for k in tokens if k not in stop_words])
data_useful
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
data_useful.loc[:,'title_tokens'] = data_useful['title_tokens'].apply(lambda tokens: [lemmatizer.lemmatize(k) for k in tokens])
data_useful.loc[:,'description_tokens'] = data_useful['description_tokens'].apply(lambda tokens: [lemmatizer.lemmatize(k) for k in tokens])
data_useful.loc[:,'content_tokens'] = data_useful['content_tokens'].apply(lambda tokens: [lemmatizer.lemmatize(k) for k in tokens])
data_useful
```

```python
data_useful.to_csv('../data/cleaned_news.csv', index=False)
```
